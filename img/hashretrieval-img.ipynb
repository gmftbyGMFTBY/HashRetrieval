```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#柱高信息
plt.yscale('log')
Y = [5.9, 34, 16, 72]
Y1 = [320, 1327, 1166, 4953]
X = np.arange(len(Y))

bar_width = 0.3
tick_label = ['E-Commerce','Douban', 'Zh50w','LCCC']

#显示每个柱的具体高度
for x, y in zip(X,Y):
    plt.text(x+0.005,y+0.005,'%.2f' %y, ha='center',va='bottom')

for x, y1 in zip(X,Y1):
    plt.text(x+0.3,y1+0.005,'%.2f' %y1, ha='center',va='bottom')
    
# show bar
plt.bar(X, Y, bar_width, align="center", color="red", label="Database storage", alpha=0.5)
plt.bar(X+bar_width, Y1, bar_width, color="purple", align="center", label="Index storage", alpha=0.5)

plt.xlabel("Datasets")
plt.ylabel("Storage Occupation (Mb)")
plt.title('Dense Storage Occupation')

plt.xticks(X+bar_width/2, tick_label)
#显示图例
plt.legend()
# plt.show()
plt.savefig('dense.eps', format='eps', dpi=1000)
```
